```python

from logging import exception

class BalanceException(Exception):
    pass

class BankAccount:
    def __init__(self, initialAmount, acctName):
        self.balance = initialAmount
        self.name = acctName
        print(f"\nAccount '{self.name}' created.\nBalance = ${self.balance:.2f}")

    def getBalance(self):
        print(f"\nAccount '{self.name}' Balance = ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance = self.balance + amount
        print("\nDeposit complete.")
        self.getBalance()

    def viableTransaction(self, amount):
        if self.balance >= amount:
            return
        else:
            raise BalanceException(
                f"\nSorry, account '{self.name}' only has a balance of ${self.balance:.2f}"
            )

    def withdraw(self, amount):
        try:
            self.viableTransaction(amount)
            self.balance = self.balance - amount
            print('\nWithdraw complete.')
            self.getBalance()
        except BalanceException as error:
            print(f'\nWithdraw interrupted: {error}')

    def transfer(self, amount, account):
      try:
        print('\n************\n\nBegining Transfer.. ')
        self.viableTransaction(amount)
        self.balance = self.balance - amount  # Deduct the transferred amount from the source account
        account.deposit(amount)
        print('\nTransfer complete!  \n\n************')
      except BalanceException as error:
        print(f'\nTransfer interrupted. {error}')
```

```python

class InterestRewardsAcct(BankAccount):
  def deposit(self,amount):
    self.balance=self.balance+(amount*1.05)
    print('\nDeposit complete.')
    self.getBalance()
```

```python

class SavingAcc(InterestRewardsAcct):
  def __init__(self, initialAmount, acctName):
    super().__init__(initialAmount, acctName)
    self.fee = 5

  def withdraw(self, amount):
    try:
      self.viableTransaction(amount + self.fee)
      self.balance = self.balance - (amount + self.fee)
      print('\nWithdraw completed.')
      self.getBalance()
    except BalanceException as error:
      print(f'\nWithdraw interrupted: {error}')
```

```python
while True:
    Dave = BankAccount(1000, 'Dave')
    Sara = BankAccount(2000, 'Sara')
    Jim = InterestRewardsAcct(3000, 'Jim')
    Blaze = SavingAcc(4000, 'Blaze')

    # Take user input for account name
    input_name = input("Please enter the account name (Dave or Sara or Jim(can't transfer Money) or Blaze(can't Transfer Money)), or type 'exit' to quit: ")

    if input_name.lower() == 'exit':
        break  # Exit the loop if the user types 'exit'
    elif input_name == 'Dave':
        account = Dave
    elif input_name == 'Sara':
        account = Sara
    elif input_name == 'Jim':
        account = Jim
    elif input_name == 'Blaze':
        account = Blaze
    else:
        print('Invalid account name. Try again.')
        continue  # Go back to the beginning of the loop if the account name is invalid

    if isinstance(account, (BankAccount, InterestRewardsAcct, SavingAcc)):
        # Take user input for the service
        while True:
            print("What kind of service do you want:")
            print("1. Deposit\n2. Withdraw\n3. Transfer")
            try:
                input_service = int(input())

                if 1 <= input_service <= 3:
                    break  # Exit the service loop if the input is valid
                else:
                    print('Invalid service. Please enter a number between 1 and 3.')
            except ValueError:
                print('Invalid input. Please enter a number.')

        if input_service == 1:
            amount = int(input("Enter the deposit amount: "))
            account.deposit(amount)
        elif input_service == 2:
            amount = int(input("Enter the withdrawal amount: "))
            account.withdraw(amount)
        elif input_service == 3:
            amount = int(input("Enter the transfer amount: "))
            target_account_name = input("Enter the target account name: ")
            target_account = Dave if target_account_name == 'Dave' else Sara
            account.transfer(amount, target_account)

        # Display updated balance
        print(f"{account.name}'s Balance: ${account.balance:.2f}")
    else:
        print('Invalid account type. Try again.')
```
